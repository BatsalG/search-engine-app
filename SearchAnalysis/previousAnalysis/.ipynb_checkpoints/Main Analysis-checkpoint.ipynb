## Data Analysis Pipeline
### Based on the data collected from the Google and Bing scrapers, this notebook is capable of analyzing the sentiments, and the political leaning of the publishers that are most represented in the search engines.

#### Note: These results might not be replicated unless performed with the exact data. The results from Google and Bing are frequently updated, and hence every time a new CSV is used, the results will vary.

This notebook creates graphs that compares the document sentiment, the emotion affinities, and the political leaning of the articles from Google and Bing News, for the 30 keywords for which the data is collected.

```python
# Import the necessary libraries.

import pandas as pd
import math
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from collections import defaultdict
```

### Get the data from Google and Bing CSVs that are created from the prior notebooks.
There are a few info that we obtain from the IBM's Sentiment Analysis:
* Total Characters
* Overall document sentiment
* 5 Emotions conveyed by the entire document
    * Sadness
    * Joy
    * Fear
    * Disgust
    * Anger

```python
# Read the data from Google and Bing CSVs.
google_data = pd.read_csv('google_with_sentiment.csv')
bing_data = pd.read_csv('bing_with_sentiment.csv')
```

```python
# Extract only the needed data for further analysis.
google_sentiment_data = google_data[['keyword','total_chars', 'doc_sentiment_num', 'sadness', 'joy', 'fear', 'disgust', 'anger']]
bing_sentiment_data = bing_data[['keyword','total_chars', 'doc_sentiment_num', 'sadness', 'joy', 'fear', 'disgust', 'anger']]
```

```python
# Group the sentiments by the keywords.
data_by_keyword_google = google_sentiment_data.groupby(
    'keyword'
).mean().round(3)

data_by_keyword_bing = bing_sentiment_data.groupby(
    'keyword'
).mean().round(3)
```

```python
# Create a complete DataFrame that includes all the info about the sentiments.
chars_sentiment_google_bing = pd.DataFrame({
    'Google Characters': data_by_keyword_google['total_chars'],
    'Bing Characters': data_by_keyword_bing['total_chars'],
    'Google Sentiment': data_by_keyword_google['doc_sentiment_num'],
    'Bing Sentiment': data_by_keyword_bing['doc_sentiment_num'],
    'Google Sadness': data_by_keyword_google['sadness'],
    'Bing Sadness': data_by_keyword_bing['sadness'],
    'Google Joy': data_by_keyword_google['joy'],
    'Bing Joy': data_by_keyword_bing['joy'],
    'Google Fear': data_by_keyword_google['fear'],
    'Bing Fear': data_by_keyword_bing['fear'],
    'Google Disgust': data_by_keyword_google['disgust'],
    'Bing Disgust': data_by_keyword_bing['disgust'],
    'Google Anger': data_by_keyword_google['anger'],
    'Bing Anger': data_by_keyword_bing['anger']
})

chars_sentiment_google_bing = chars_sentiment_google_bing.round(3)
num_of_rows = len(chars_sentiment_google_bing.index)
```

```python
# Compare the average sentiments and emotions for the articles of Bing and Google.
google_great_bing_sentiment = 0
google_great_bing_article_length = 0
google_great_bing_sadness = 0
google_great_bing_joy = 0
google_great_bing_fear = 0
google_great_bing_disgust = 0
google_great_bing_anger = 0

# Keep Track of the sentiments, and also how many topics have a higher or lower sentiment value across the search engines.
for row in range(num_of_rows):
    diff_sentiment = (chars_sentiment_google_bing['Google Sentiment'][row] > chars_sentiment_google_bing['Bing Sentiment'][row])
    diff_length = (chars_sentiment_google_bing['Google Characters'][row] > chars_sentiment_google_bing['Bing Characters'][row])
    diff_sadness = (chars_sentiment_google_bing['Google Sadness'][row] > chars_sentiment_google_bing['Bing Sadness'][row])
    diff_joy = (chars_sentiment_google_bing['Google Joy'][row] > chars_sentiment_google_bing['Bing Joy'][row])
    diff_fear = (chars_sentiment_google_bing['Google Fear'][row] > chars_sentiment_google_bing['Bing Fear'][row])
    diff_disgust = (chars_sentiment_google_bing['Google Disgust'][row] > chars_sentiment_google_bing['Bing Disgust'][row])
    diff_anger = (chars_sentiment_google_bing['Google Anger'][row] > chars_sentiment_google_bing['Bing Anger'][row])
    if diff_sentiment == True:
        google_great_bing_sentiment += 1
    if diff_length == True:
        google_great_bing_article_length += 1
    if diff_sadness == True:
        google_great_bing_sadness += 1
    if diff_joy == True:
        google_great_bing_joy += 1
    if diff_fear == True:
        google_great_bing_fear += 1
    if diff_disgust == True:
        google_great_bing_disgust += 1
    if diff_anger == True:
        google_great_bing_anger += 1
        
bing_great_google_sentiment = num_of_rows - google_great_bing_sentiment
bing_great_google_article_length = num_of_rows - google_great_bing_article_length
bing_great_google_sadness = num_of_rows - google_great_bing_sadness
bing_great_google_joy = num_of_rows - google_great_bing_joy
bing_great_google_fear = num_of_rows - google_great_bing_fear
bing_great_google_disgust = num_of_rows - google_great_bing_disgust
bing_great_google_anger = num_of_rows - google_great_bing_anger
```

### Get and tabulate the most frequenly occuring publishers across Google and Bing.

```python
# Finds the top publisher based on the data provided.
def get_top_publisher(data, from_, to_):
    all_publisher = {}
    top_publisher = {}
    
    # Go through all the data, clean the data for manual comparison.
    # Creates a dict with the top 10 publishers.
    for i in data:
        publisher = i
        if (i[-11:-1]) == ' on MSN.co':
            publisher = i[:-11]
        if publisher == "New York Times":
            publisher = "The New York Times"
        if publisher in all_publisher:
            all_publisher[publisher] += 1
        else:
            all_publisher[publisher] = 1
    top_raw = sorted(all_publisher, key=all_publisher.get, reverse=True)[from_:to_]
    for i in top_raw:
        top_publisher[i] = all_publisher[i]
    return top_publisher
```

```python
top_publisher_google = get_top_publisher(google_data["publisher"], 0, 10)
top_publisher_bing = get_top_publisher(bing_data["publisher"], 0, 11)

df_top_google_publisher = pd.DataFrame(
    {
        'Google Publisher': top_publisher_google.keys(),
        'Count': top_publisher_google.values()
    })

df_top_bing_publisher = pd.DataFrame(
    {'Bing Publisher': top_publisher_bing.keys(),
     'Count': top_publisher_bing.values()
    })
```

### AllSides Data
AllSides is a popular platform that ranks the publishers political leaning based on public perception.

Since, AllSides does not have an API that makes it easy to fetch data, all the rankings are downloaded as a CSV and stored in the working directory. It is named as 'allsides_data.csv'.

```python
allsides_raw_data = pd.read_csv('./allsides_data.csv')
allsides_data = pd.DataFrame(
    {
        'publisher': allsides_raw_data['news_source'],
        'rating': allsides_raw_data['rating'],
        'agree': allsides_raw_data['perc_agree']
    })
allsides_data
```

Not all publishers from Google and Bing are present in the data from AllSides, and the names of the publishers are oftentimes different. So, we match the data from AllSides to the one obtained through scraping Bing and Google.

**Note**: There might not always be data for all the publishers. In the graph, they will be bars that are grey.

```python
def political_leaning(df_top_publisher, allsides_data, txt):
    politics_rating = []
    politics_agree = []
    politics_pub = []
    
    for i in range(len(df_top_publisher[txt])):
        pub = df_top_publisher[txt][i]
        pub = pub.lower()
        if pub[:4] == "the ":
            pub = pub[4:]

        for all_publisher in range(len(allsides_data['publisher'])):
            as_pub = (allsides_data['publisher'][all_publisher]).lower()
            if pub in as_pub:
                politics_rating.append(allsides_data['rating'][all_publisher])
                politics_agree.append(allsides_data['agree'][all_publisher])
                politics_pub.append([as_pub, pub])
                break
            elif as_pub in pub:
                politics_rating.append(allsides_data['rating'][all_publisher])
                politics_agree.append(allsides_data['agree'][all_publisher])
                politics_pub.append([as_pub, pub])
                break
            else:
                if (as_pub[-4:-1] == '.co'):
                    temp_as = as_pub[:-4]
                    temp = pub.replace(" ", "")
                    if temp_as in temp:
                        politics_rating.append(allsides_data['rating'][all_publisher])
                        politics_agree.append(allsides_data['agree'][all_publisher])
                        politics_pub.append([as_pub, pub])
                        break
        if len(politics_rating) != i+1:
            politics_rating.append('')
            politics_agree.append('')
            politics_pub.append('')
    
    return politics_rating, politics_agree, politics_pub
```

```python
google_rating, google_politics_agree, google_politics_pub = political_leaning(df_top_google_publisher, allsides_data, 'Google Publisher')
bing_rating, bing_politics_agree, bing_politics_pub = political_leaning(df_top_bing_publisher, allsides_data, 'Bing Publisher')
```
